```python
import pandas as pd
```

```python
df_atari = pd.read_csv('Muj_PPO.csv')
df_atari = df_atari[['Environment', 'Algo','Seed', 'Best Mean Reward']]
df_atari_a2c = df_atari[df_atari['Algo'] == 'PPO']
```

```python
df_atari_a2c
```

df_atari_a2c = df_atari_a2c.groupby('Environment')['Best Mean Reward'].mean().round(2).reset_index()
df_atari_a2c = df_atari_a2c.rename(columns={'Best Mean Reward': 'Score'}) 

```python
# Group by 'Environment' and count occurrences
environment_counts = df_atari_a2c['Environment'].value_counts()

# Filter environments with exactly 4 occurrences
valid_environments = environment_counts[environment_counts == 4].index

# Merge to keep only valid environments
filtered_df_atari_a2c = df_atari_a2c.merge(
    pd.DataFrame({'Environment': valid_environments}), on='Environment'
)

# Group by 'Environment' and calculate mean of 'Best Mean Reward'
filtered_df_atari_a2c = (
    filtered_df_atari_a2c.groupby('Environment')['Best Mean Reward']
    .mean()
    .round(2)
    .reset_index()
)
```
